```python
for i in range(len(df)):
    #Assign the current team to a temp value of home team, away team and result.
    HT = df.iloc[i][2]
    AT = df.iloc[i][3]
    res = df.iloc[i][6]

    if res == "H":
        team_sum.loc[HT]["Win"] += 1
        team_sum.loc[AT]["Loss"] += 1
    elif res == "A":
        team_sum.loc[AT]["Win"] += 1
        team_sum.loc[HT]["Loss"] += 1
    else:
        team_sum.loc[HT]["Draw"] += 1
        team_sum.loc[AT]["Draw"] += 1
```

```python
team_sum.to_csv('wins_per_team.csv', index=True)
```

```python
team_sum["%W"] = 0
```

```python
team_sum
```

```python
for i in range (len(team_sum)):
    
    GP = team_sum.iloc[i]["GP"]
    wins = team_sum.iloc[i]["Win"]
    win_frac = (wins / GP)
    win_percent = "{:.2%}".format(win_frac)

    team_sum.iat[i, 4] = win_percent
```

```python
team_sum
```

```python
for i in range (len(team_sum)):
    
    GP = team_sum.iloc[i]["GP"]
    draw = team_sum.iloc[i]["Draw"]
    draw_frac = (draw / GP)
    draw_percent = "{:.2%}".format(draw_frac)

    team_sum.iat[i, 5] = draw_percent
```

```python
for i in range (len(team_sum)):
    
    GP = team_sum.iloc[i]["GP"]
    loss = team_sum.iloc[i]["Loss"]
    loss_frac = (loss / GP)
    loss_percent = "{:.2%}".format(loss_frac)

    team_sum.iat[i, 6] = loss_percent
```
